## Making of RUL ( Remaining useful life ) column

```python
#RUL= Max cycle time - current cycle
max_cycle= data.groupby(['engine no.'],sort=False)['cycles'].max().reset_index().rename(columns={'cycles':'max cycle'})
max_cycle.head()
```

```python
#merging the data 
new_data=pd.merge(data,max_cycle)
new_data["RUL"]=new_data['max cycle']-new_data['cycles']
new_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

# Data Visualisation 

```python
plt.bar(data['engine no.'] ,data['cycles'])
plt.xlabel('engine number')
plt.ylabel('cycles')
plt.show()
```

```python
import math
plt.subplots(figsize=(12, 6))
sns.histplot(new_data["max cycle"],kde=True,  color='blue');
print("Mean number of cycles after which jet engine fails is "+ str(math.floor(new_data["max cycle"].mean())))
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(data.corr(),annot=True,cmap='Blues',linewidths=1)
plt.show()
```

```python
new_data.nunique(axis=0)
# columns which dont have unique values we can drop them
```

# Box plot

```python
plt.figure(figsize=(15,21))
for i in np.arange (1,25):
    temp=new_data.iloc[:,i]
    plt.subplot(5,6,i)
    plt.boxplot(temp)
    plt.title(new_data.columns[i])
plt.show()
```

# classification problem formulation 

```python
# life ratio = cycles/RUL
# if LR <=0.6 = good condition , does not required maintenance 
#0.6<LR =<0.8 = moderate condition ,required maintenance
#LR>0.8= warning condition 
#labels ,good condition=0
#moderate condition=1
#warning condition=2
new_data['LR']=new_data['cycles'].div(new_data['RUL'])
new_data.drop(['RUL','max cycle'],axis=1,inplace=True)
new_data.head()
```

```python
labels=[]
for i in range (0,len(new_data)):
    if np.array(new_data['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array (new_data['LR'])[i]<= 0.8:
        labels.append(1)
    else:
        labels.append(2)
new_data['labels']=labels
new_data.drop(['LR'],axis=1,inplace=True)
new_data.head()
```

```python
new_data.tail()
```

```python
new_data.to_csv('Training 1 data.csv' , index=None)
```

# merging all 4 CSV files

```python
df1=pd.read_csv('Training 1 data.csv')
df2=pd.read_csv('Training 2 data.csv')
df3=pd.read_csv('Training 3 data.csv')
df4=pd.read_csv('Training 4 data.csv')
```

```python
df=pd.concat([df1,df2,df3,df4])
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X=df.drop('labels',axis=1)
y=df['labels']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20 ,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred2=dt.predict(X_test)
accuracy_score(y_test,y_pred2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rm=RandomForestClassifier()
rm.fit(X_train,y_train)
y_pred3=rm.predict(X_test)
accuracy_score(y_test,y_pred3)
```

```python
 final_data=pd.DataFrame({'models':['LR','DT','RM'],'Accuracy':[accuracy_score(y_test,y_pred1)*100,
                                                    accuracy_score(y_test,y_pred2)*100,
                                                   accuracy_score(y_test,y_pred3)*100]})
```

```python
sns.barplot(final_data['models'],final_data['Accuracy'])
plt.show()
```
